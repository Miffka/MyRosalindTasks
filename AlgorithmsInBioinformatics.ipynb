```python
# Task 3.5.5 Greedy Motif Search

import numpy as np

def equalHammingDistance(word1, word2):
    return sum(map(lambda x: 1 if word1[x]!=word2[x] else 0, range(len(word1))))

def distPattStrings(pattern, strings):
    k = len(pattern)
    totalD = 0
    for text in strings:
        hammD = float('inf')
        for i in range(len(text)-k +1):
            if hammD > equalHammingDistance(pattern, text[i:i+k]):
                hammD = equalHammingDistance(pattern, text[i:i+k])
        totalD += hammD
    return totalD


# Takes list or tuple of strings
def probMatrix(strings, alphabet = {'A':0, 'C':1, 'G':2, 'T':3}):
    k = len(strings[0])
    probMatr = np.zeros((len(alphabet), k))
    for text in strings:
        for index, letter in enumerate(text):
            probMatr[alphabet[letter], index] += 1
    probMatr /= len(strings)
    return probMatr

def profMostProbKmer(text, k, probMatr, alphabet = {'A':0, 'C':1, 'G':2, 'T':3}):
    index = list()
    for i in range(len(text)-k+1):
        pat = text[i:i+k]
#        print(pat)
        patProb = np.prod(list(map(lambda x: probMatr[alphabet[x[1]],x[0]], enumerate(pat))))
        index.append([pat, patProb])
#    print(index)
    index.sort(key = lambda x: x[1], reverse = True)
    return index[0][0]

# Takes list of strings or probability matrix
def consensus(strings, probMatr = None, alphabet = ['A', 'C', 'G', 'T']):
    if not probMatr:
        probMatr = probMatrix(strings)
    return ''.join(tuple(map(lambda x: alphabet[x], np.argmax(probMatr, axis = 0))))

# Takes list of strings or probability matrix
def score(strings, alphabet = ['A', 'C', 'G', 'T']):
    cons = consensus(strings)
    return distPattStrings(cons, strings)
    
def greedyMotifSearch(strings, k, t):
    bestMotifs = list(map(lambda x: x[:k], strings))
    firstString = strings[0]
    # iterate for k-mers in the first string
    for i in range(len(firstString)-k+1):
        motifs = list()
        motifs.append(firstString[i:i+k])
        # walk through other strings
        for j in range(1, t):
            profile = probMatrix(motifs)
            motifs.append(profMostProbKmer(strings[j], k, profile))
#        print(motifs)
#        print(bestMotifs)
#        print(score(motifs), score(bestMotifs))
        if score(motifs) < score(bestMotifs):
            bestMotifs = motifs
    return bestMotifs
            
    


with open('dataset_159_5.txt') as inf:
#    inf.readline()
    k, t = map(int, inf.readline().split())
    strings = list()
    for i in range(t):
        strings.append(inf.readline().strip())

#print(probMatrix(strings))
#print(consensus(strings))
#print('\n'.join(strings))
print(greedyMotifSearch(strings, k, t))
```

```python
# Task 3.6.9 Greedy motif search with pseudocounts
# We should only substitute our profile() function

import numpy as np

def equalHammingDistance(word1, word2):
    return sum(map(lambda x: 1 if word1[x]!=word2[x] else 0, range(len(word1))))

def distPattStrings(pattern, strings):
    k = len(pattern)
    totalD = 0
    for text in strings:
        hammD = float('inf')
        for i in range(len(text)-k +1):
            if hammD > equalHammingDistance(pattern, text[i:i+k]):
                hammD = equalHammingDistance(pattern, text[i:i+k])
        totalD += hammD
    return totalD

# Takes list or tuple of strings
def probMatrix(strings, alphabet = {'A':0, 'C':1, 'G':2, 'T':3}):
    k = len(strings[0])
    probMatr = np.zeros((len(alphabet), k))
    for text in strings:
        for index, letter in enumerate(text):
            probMatr[alphabet[letter], index] += 1
    probMatr /= len(strings)
    return probMatr

# Takes list of strings
def probMatrPseudo(strings, alphabet = {'A':0, 'C':1, 'G':2, 'T':3}):
    k = len(strings[0])
    probMatr = np.zeros((len(alphabet), k))
    for text in strings:
        for index, letter in enumerate(text):
            probMatr[alphabet[letter], index] += 1
    # Here we add one to each section
    probMatr += 1
    probMatr /= len(strings)+4
    return probMatr

# Takes one-line text, k for the length of k-mer and probability matrix
def profMostProbKmer(text, k, probMatr, alphabet = {'A':0, 'C':1, 'G':2, 'T':3}):
    index = list()
    for i in range(len(text)-k+1):
        pat = text[i:i+k]
        patProb = np.prod(list(map(lambda x: probMatr[alphabet[x[1]],x[0]], enumerate(pat))))
        index.append([pat, patProb])
    index.sort(key = lambda x: x[1], reverse = True)
    return index[0][0]

# Takes list of strings or probability matrix
def consensus(strings, probMatr = None, alphabet = ['A', 'C', 'G', 'T']):
    if not probMatr:
        probMatr = probMatrix(strings)
    return ''.join(tuple(map(lambda x: alphabet[x], np.argmax(probMatr, axis = 0))))

# Takes list of strings or probability matrix
def score(strings, alphabet = ['A', 'C', 'G', 'T']):
    cons = consensus(strings)
    return distPattStrings(cons, strings)
    
def greedyMotifSearch(strings, k, t):
    bestMotifs = list(map(lambda x: x[:k], strings))
    firstString = strings[0]
    # iterate for k-mers in the first string
    for i in range(len(firstString)-k+1):
        motifs = list()
        motifs.append(firstString[i:i+k])
        # walk through other strings
        for j in range(1, t):
            profile = probMatrPseudo(motifs)
            motifs.append(profMostProbKmer(strings[j], k, profile))
        if score(motifs) < score(bestMotifs):
            bestMotifs = motifs
    return bestMotifs


with open('dataset_160_9.txt') as inf:
#    inf.readline()
    k, t = map(int, inf.readline().split())
    strings = list()
    for i in range(t):
        strings.append(inf.readline().strip())

#print(probMatrix(strings))
#print(consensus(strings))
#print('\n'.join(strings))
print('\n'.join(greedyMotifSearch(strings, k, t)))
```
